#### Implementation of Support Vector Classifiers

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Lets Create synthetic data points
from sklearn.datasets import make_classification

x,y = make_classification(n_samples = 1000,n_features = 2,n_classes = 2,n_clusters_per_class = 2,n_redundant = 0 )
```

```python
print(x)
print(y)
```

```python
pd.DataFrame(x)[0]
```

```python
sns.scatterplot(x = pd.DataFrame(x)[0],y = pd.DataFrame(x)[1],hue = y)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC

svc = SVC(kernel = 'linear')
svc.fit(x_train,y_train)
```

```python
svc.coef_
```

```python
y_pred = svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(" Classification Report : ",classification_report(y_test,y_pred))
print("\nConfusion Matrix : ",confusion_matrix(y_test,y_pred))
```

```python
from sklearn.svm import SVC

rbf = SVC(kernel = 'rbf')
rbf.fit(x_train,y_train)
```

```python
y_pred1 = rbf.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(" Classification Report : ",classification_report(y_test,y_pred1))
print("\nConfusion Matrix : ",confusion_matrix(y_test,y_pred1))
```

```python
polynomial = SVC(kernel = 'poly')
polynomial.fit(x_train,y_train)
```

```python
y_pred2 = polynomial.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(" Classification Report : ",classification_report(y_test,y_pred2))
print("\nConfusion Matrix : ",confusion_matrix(y_test,y_pred2))
```

```python
sigmoid = SVC(kernel = 'sigmoid')
sigmoid.fit(x_train,y_train)
```

```python
y_pred3 = sigmoid.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(" Classification Report : ",classification_report(y_test,y_pred3))
print("\nConfusion Matrix : ",confusion_matrix(y_test,y_pred3))
```

#### HyperParameter Tuning with SVC

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Defining parameter range
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf']
}

grid = GridSearchCV(
    SVC(),
    param_grid=param_grid,
    refit=True,
    cv=5,
    verbose=3
)

grid.fit(x_train, y_train)

```

```python
grid.best_params_
```

```python
y_pred4 = grid.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(" Classification Report : ",classification_report(y_test,y_pred4))
print("\nConfusion Matrix : ",confusion_matrix(y_test,y_pred4))
```

c = Regularization Parameter

gamma = Influence of a single data point

Total Combination :
5(c) * 5(gamma) * 1(kernel) = 25 models

Note : Here we used GridSearchCV to perform Hyperparameter tuning.

It checks all combos of given paameters using cv and selects the model with best performance....
